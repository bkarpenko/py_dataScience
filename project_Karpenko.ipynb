```python
# определяем среднюю площадь квартиры в зависимости от кол комнат
a = data.loc[(data.Square > data.LifeSquare - 15) & (data.LifeSquare > 20)]

dataByRooms = a.groupby(["Rooms"], as_index=False)[["Square", "LifeSquare", "Price"]].mean()
dataByRooms["LifeSquareShare"] = dataByRooms.LifeSquare / dataByRooms.Square
dataByRooms
```

```python
data.Rooms.value_counts()
```

```python
# корректируем кол комнат для квартир с аномалиями
data.loc[data.Rooms == 6, "Rooms"] = 2
data.loc[data.Rooms == 10, "Rooms"] = 2
data.loc[data.Rooms == 19, "Rooms"] = 1
```

```python
data[data.Rooms == 0]
```

```python
# удаляем строки, где кол комнат = 0
data.drop(data[data.Rooms == 0].index, axis = 0, inplace = True)
```

```python
data.describe()
```

```python
#корректируем Square
def prepare_square(df):
    df.loc[df['Square']>(df['Rooms']*40+40),'Square']=df['Rooms']*40+40
    df.loc[df['Square']<(df['Rooms']*15),'Square']=df['Rooms']*15
    return df
```

```python
data=prepare_square(data)
```

```python
data.describe()
```

```python
# добавляем столбец со средней долей жилой площади от общей
data = pd.merge(data, dataByRooms[["Rooms", "LifeSquareShare"]], on = "Rooms", how = "left")
data.head()
```

```python
# корректируем жилую площадь для квартир 

#с аномально маленьким показателем
data2 = data.copy()
data2.loc[data2.LifeSquare < (data2.Square / 3), "LifeSquare"] = data2.Square * data2.LifeSquareShare
# где жилая больше общей
data2.loc[data2.Square < data2.LifeSquare, "LifeSquare"] = data2.Square * data2.LifeSquareShare
# где na
data2.loc[np.isnan(data2.LifeSquare), "LifeSquare"] = data2.Square * data2.LifeSquareShare
```

```python
def prepare_lifesquare(df):
    df.loc[df['Square']<df['LifeSquare'],'LifeSquare']=df['Square']
    return df
```

```python
data = prepare_lifesquare(data)
```

```python
def fillna_life_square(df,df_source):
    df['LifeSquare']=df['LifeSquare'].fillna(df_source['LifeSquare'].mean())
    return df
```

```python
data = fillna_life_square(data,data)
```

```python
def join_stats1(df,stats1,source_df,mode='train'):
    df=pd.merge(df,stats1,on=['DistrictId','Rooms'],how='left')
    if mode=='test':
        df['mean_price']=df['mean_price'].fillna(source_df['mean_price'].mean())
    return df
```

```python
dataMeanPrice=data.groupby(['DistrictId','Rooms'],as_index=False)[['Price']].mean().rename(columns={'Price':'mean_price'})
```

```python
dataMeanPrice.head()
```

```python
data = join_stats1(data, dataMeanPrice, data)
data2 = join_stats1(data2, dataMeanPrice, data2)
```

присвоим порядковый номер району исходя из стоимости за квартиру

```python
data["PriceMeter"] = data.Price / data.Square
data2["PriceMeter"] = data2.Price / data2.Square
```

```python
sample = data.groupby(["DistrictId"],as_index=False)[['PriceMeter']].mean()
sample.sort_values("PriceMeter", inplace = True)
sample["PriceId"] = list(range(len(sample)))
```

```python
sample.head()
```

```python
# добавляем столбец с порядковым номером района исходя из стоимости за квартиру
data = pd.merge(data, sample[["DistrictId", "PriceId"]], on = "DistrictId", how = "left")
data2 = pd.merge(data2, sample[["DistrictId", "PriceId"]], on = "DistrictId", how = "left")
```

```python
data2.describe()
```

```python
features1=data[['Rooms','Square','LifeSquare','mean_price']]
target1=data['Price']
```

```python
features2=data2[['Rooms','Square','LifeSquare','mean_price', 'Floor', "HouseYear"]]
target2=data2['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression as LR, Ridge
from sklearn.ensemble import RandomForestRegressor as RF
from sklearn.metrics import mean_squared_error as mse,r2_score as r2
```

```python
lr_model= LR()
```

LinearRegression 1

```python
lr_model.fit(X_train1,y_train1)
y_pred1 = lr_model.predict(X_test1)
print(mse(y_test1,y_pred1),r2(y_test1,y_pred1))
```

LinearRegression 2

```python
lr_model.fit(X_train2,y_train2)
y_pred2 = lr_model.predict(X_test2)
print(mse(y_test2,y_pred2),r2(y_test2,y_pred2))
```

Ridge 2
